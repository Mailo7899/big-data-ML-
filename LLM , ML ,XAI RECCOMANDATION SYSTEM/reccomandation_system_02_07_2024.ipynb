#elaboro dati :embeddings e tf-idf

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from rapidfuzz import process
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.metrics.pairwise import cosine_similarity
# Caricamento dei dati (supponendo che df_cleaned_rating_unito2 sia già caricato)

# Embedding del testo con SBERT
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings = model.encode(df_cleaned_rating_unito2['description'], show_progress_bar=True)

import torch
#print("GPU disponibile:", torch.cuda.is_available())

# Importa le librerie necessarie
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics.pairwise import cosine_similarity
# Configura il dispositivo per utilizzare la GPU se disponibile
#device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#print(f"Utilizzando il dispositivo: {device}")

# Calcolo della similarità
cosine_sim = cosine_similarity(embeddings, embeddings)



from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Calcolo della matrice TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df_cleaned_rating_unito2['description'])






# Ridurre la dimensionalità della matrice TF-IDF
svd = TruncatedSVD(n_components=100)
normalizer = Normalizer(copy=False)
lsa = make_pipeline(svd, normalizer)

tfidf_matrix_reduced = lsa.fit_transform(tfidf_matrix)

# Calcolo della similarità cosine con la matrice ridotta
cosine_sim_tfidf_reduced = cosine_similarity(tfidf_matrix_reduced)


# Convertire tutti i titoli in minuscolo
df_cleaned_rating_unito2['Title'] = df_cleaned_rating_unito2['Title'].str.lower()



```

#reccomandation system finale

```python
import gradio as gr
from rapidfuzz import process
import pandas as pd
import os

# Assicurati che df_cleaned_rating_unito2, cosine_sim e cosine_sim_tfidf_reduced siano definiti correttamente
# df_cleaned_rating_unito2 = ...
# cosine_sim = ...
# cosine_sim_tfidf_reduced = ...

# Funzioni di raccomandazione
def get_recommendations_embeddings(title, top_n=10):
    title = title.lower()
    matched_title, score, index = process.extractOne(title, df_cleaned_rating_unito2['Title'])
    sim_embeddings = cosine_sim[index]
    similar_indices = sim_embeddings.argsort()[::-1]
    recommended_books = df_cleaned_rating_unito2.iloc[similar_indices[:top_n]].copy()
    recommended_books['SimilarityEmbeddings'] = sim_embeddings[similar_indices[:top_n]]
    return recommended_books[['Title', 'authors', 'categories', 'rating_mean', 'SimilarityEmbeddings']]

def get_recommendations_tfidf(title, top_n=10):
    title = title.lower()
    matched_title, score, index = process.extractOne(title, df_cleaned_rating_unito2['Title'])
    sim_tfidf_reduced = cosine_sim_tfidf_reduced[index]
    similar_indices = sim_tfidf_reduced.argsort()[::-1]
    recommended_books = df_cleaned_rating_unito2.iloc[similar_indices[:top_n]].copy()
    recommended_books['SimilarityTFIDF'] = sim_tfidf_reduced[similar_indices[:top_n]]
    return recommended_books[['Title', 'authors', 'categories', 'rating_mean', 'SimilarityTFIDF']]

def get_combined_recommendations(title):
    recommended_books_embeddings = get_recommendations_embeddings(title)
    recommended_books_tfidf = get_recommendations_tfidf(title)
    return recommended_books_embeddings, recommended_books_tfidf

# Percorso del file per salvare il feedback
feedback_file = '/content/drive/MyDrive/Colab Notebooks/2024_BDA/data/feedback.csv'

def collect_feedback(title, feedback):
    feedback_entry = {'Title': title, 'Preferred': feedback}

    # Verifica se il file esiste già
    file_exists = os.path.isfile(feedback_file)

    try:
        # Apri il file in modalità append e scrivi solo la nuova riga
        with open(feedback_file, 'a') as f:
            df = pd.DataFrame([feedback_entry])
            df.to_csv(f, header=not file_exists, index=False)
        print("Feedback saved successfully.")
    except Exception as e:
        print(f"An error occurred while saving the feedback: {e}")

    return None  # Reset Radio button

# Definisce l'interfaccia di Gradio
with gr.Blocks() as demo:
    gr.Markdown("## Sistema di Raccomandazione di Libri")
    input_text = gr.Textbox(label="Inserisci il titolo di un libro")
    output_embeddings = gr.DataFrame(label="Raccomandazioni basate su Embeddings")
    output_tfidf = gr.DataFrame(label="Raccomandazioni basate su TF-IDF")
    feedback = gr.Radio(choices=["Embeddings", "TF-IDF"], label="Quale output preferisci? Una volta selezionato l'output che preferisci, clicca nuovamente su 'Invia'")
    submit_button = gr.Button("Invia")

    submit_button.click(fn=get_combined_recommendations, inputs=input_text, outputs=[output_embeddings, output_tfidf])
    submit_button.click(fn=collect_feedback, inputs=[input_text, feedback], outputs=feedback)

# Avvia l'interfaccia
demo.launch()
```
